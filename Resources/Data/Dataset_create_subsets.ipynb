```python
ecobici_rides_df_test = ecobici_rides_df.dropna().copy()
ecobici_rides_df_test
```

# Create manageable data sets

- oneInTen (10% the size of the original database. Takes every tenth row
- oneInHundred (1% the size of the original database. Takes every hundredth row

```python
oneInTen = []
lngth = len(ecobici_rides_df_test)
for i in range(lngth):
    if i%10==0: oneInTen.append(i)
oneInTen

ecobici_rides_df_oneInTen = ecobici_rides_df_test.iloc[oneInTen, :].copy()

ecobici_rides_df_oneInTen
```

```python
oneInHundred = []
lngth = len(ecobici_rides_df_test)
for i in range(lngth):
    if i%100==0: oneInHundred.append(i)

ecobici_rides_df_oneInHundred = ecobici_rides_df_test.iloc[oneInHundred, :].copy()

ecobici_rides_df_oneInHundred
```

```python
# Export clean CSV file to master CSV file

tic = timeit.default_timer()                            # Monitor performance

ecobici_rides_df_oneInTen.to_csv("Gitignore-data-output-files/ecobici_ride_data-oneInTen.csv", index=False) 

toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to export unified CSV file: {round(toc - tic, 2)}')

ecobici_rides_df_oneInTen.head()
```

```python
# Export clean CSV file to master CSV file

tic = timeit.default_timer()                            # Monitor performance

ecobici_rides_df_oneInHundred.to_csv("dataset_export_files/ecobici_ride_data-oneInHundred.csv", index=False) 

toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to export unified CSV file: {round(toc - tic, 2)}')

ecobici_rides_df_oneInHundred.head()
```
