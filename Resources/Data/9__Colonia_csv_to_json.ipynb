```python
# Import CSV

tic = timeit.default_timer()                            # Monitor performance

colonias_df = pd.read_csv("Routes/coloniascdmxv2.csv", index_col='id', header=0)

toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to import CSV file: {round(toc - tic, 2)}')
colonias_df
```

```python
colonias_df_test= colonias_df.copy()
```

```python
colonias_df_test.info()
```

```python
colonias_df_test.drop(colonias_df_test.iloc[:,5:], inplace=True, axis=1)
colonias_df_test

```

```python
colonias_df_test.drop(['entidad', 'geo_point_2d'], inplace=True, axis=1)
colonias_df_test
```

```python
for index, row in colonias_df_test.iterrows():
    if type(row['geo_shape']) != str:
        print("We have an imposter!")
        colonias_df_test.drop(index, inplace=True, axis=0)
colonias_df_test
```

```python
colonias_df_test['geo_shape'] = colonias_df_test['geo_shape'].map(lambda x: x.lstrip('{"type": "Polygon", "coordinates": ').rstrip('}'))
colonias_df_test
```

```python
colonias_df_test['geo_shape'] = colonias_df_test['geo_shape'].map(lambda x: x[2:-2])
colonias_df_test
```

```python
colonias_df_test2= colonias_df_test.copy()
colonias_df_test2
```

```python
colonias_df_test2['geo_shape'] = colonias_df_test2['geo_shape'].map(lambda x: x.replace('[', '').replace(']', ''))
colonias_df_test2
```

```python
colonias_df_test2['geo_shape'] = colonias_df_test2['geo_shape'].map(lambda x: x.split(', '))
colonias_df_test2
```

```python
colonias_df_test3 = colonias_df_test2.dropna().copy()
colonias_df_test3
```

```python
def createPairs(x):
    lst = []
    while(x):
        a = float(x.pop(0))
        b = float(x.pop(0))
        lst.append([b,a])

    return lst
```

```python
for index, row in colonias_df_test3.iterrows():
    try:
        lst = createPairs(row['geo_shape'])
    #     print(lst)
        colonias_df_test3.at[index, 'geo_shape'] = lst
    except: colonias_df_test3.drop(index, axis=0, inplace=True)
colonias_df_test3
```

```python
for index, row in colonias_df_test3.iterrows():
    if type(row['geo_shape']) == list: continue
    else: colonias_df_test3.drop(index, axis=0, inplace=True)
colonias_df_test3
```

```python
colonias_df_test3.reset_index(drop=True, inplace=True)
colonias_df_test3
```
